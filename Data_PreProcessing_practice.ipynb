# Remove nan from one specific column

```python
cleandata = datasetT.dropna(subset =['LoanAmount'])
cleandata
```

```python
dt = datasetT.copy()
dt.columns
```

```python
cols = ['Gender','Area', 'Loan_Status']

dt[cols] = dt[cols].fillna(dt.mode().iloc[0])
```

```python
dt.isna().sum()
```

# Remove numerical values

```python
col2 = ['ApplicantIncome','CoapplicantIncome','LoanAmount']

dt[col2] = dt[col2].fillna(dt[col2].mean()) # note: to fill with mean we have to give 'column_name' with dataset inside'fillna'

dt.isna().sum()
```

# Label Encoding

```python
# Change object data types to 'category'
```

```python
dt.dtypes
```

```python
dt[cols] = dt[cols].astype('category')
```

```python
dt.dtypes
```

```python
for c in cols:
  dt[c] = dt[c].cat.codes
```
