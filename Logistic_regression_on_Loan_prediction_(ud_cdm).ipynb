```python
df = pd.read_csv('/content/drive/MyDrive/ud ds+ml (logistic regress....)/01Exercise1.csv')
df
```

```python
# copy then clean dataset
```

```python
LoanPrep = df.copy()
```

```python
LoanPrep.shape
```

```python
LoanPrep.isnull().sum()
```

```python
LoanPrep = LoanPrep.dropna()
```

```python
LoanPrep.shape
```

```python
# create dummy variable for categorical values.
```

```python
LoanPrep = LoanPrep.drop(['gender'] , axis = 1)
```

```python
LoanPrep.columns
```

```python
LoanPrep.head(4)
```

```python
LoanPrep[['married','status']] = LoanPrep[['married','status']].astype('category')
```

```python
LoanPrep.dtypes
```

```python
dummies = LoanPrep[['married','status']]
dummies = pd.get_dummies(dummies , drop_first= True)
dummies = dummies.astype('int')
LoanPrep = LoanPrep.drop(['married','status'] , axis= 1)
LoanPrep = pd.concat([dummies , LoanPrep] , axis= 1)
```

```python
LoanPrep
```

```python
# Normalize the data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar_ = StandardScaler()
```

```python
LoanPrep['income'] = scalar_.fit_transform(LoanPrep[['income']])
LoanPrep['loanamt'] = scalar_.fit_transform(LoanPrep[['loanamt']])
```

```python
LoanPrep
```

```python
# Create the X and Y
```

```python
y = LoanPrep['status_Y']
X = LoanPrep.drop(['status_Y'] , axis = 1)
```

```python
# split for train and test
```

```python
X_train , X_test , y_train , y_test = train_test_split( X , y , test_size= .3 , random_state= 42 , stratify= y)
```

```python
# Build the logistic regression Model
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(X_train , y_train)
```

```python
y_preds = lr.predict(X_test)
```

```python
# Build confusion matrix.
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test , y_preds)
```
