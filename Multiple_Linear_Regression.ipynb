# MULTIPLE LINEAR REGRESSION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
import statistics as st

# Libraries For Text

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download('punkt')
nltk.download('stopwords')



from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import confusion_matrix
%matplotlib inline
```

```python
# Read the files
```

```python
dt = pd.read_csv('/content/drive/MyDrive/Udemy ds course datasets/hours and marks/02Students.csv')
df = dt.copy()
```

```python
dt
```

```python
# Split dataset to X and Y
```

```python
X = df.iloc[: , :-1]
y = df.iloc[: , -1]
```

```python
# Train test split

from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X, y , test_size= .3 , random_state= 42 )
```

```python
# Model import and train
from sklearn.linear_model import LinearRegression

Regressore = LinearRegression()
Regressore.fit(X_train , y_train)

# test

y_preds = Regressore.predict(X_test)
y_preds

# Create R-Squired value

Mlr_score = Regressore.score(X_test , y_test)
Mlr_score
```

```python
coif = Regressore.coef_
coif
```

```python
interc = Regressore.intercept_
interc
```

```python
from sklearn.metrics import mean_squared_error
import math
```

```python
rmse = math.sqrt(mean_squared_error(y_test , y_preds))

rmse
```
