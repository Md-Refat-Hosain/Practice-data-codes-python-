```python
dt = pd.read_csv('/content/drive/MyDrive/Udemy ds course datasets/hours and marks/01Students.csv')
```

```python
df = dt.copy()
```

```python
df
```

```python
# Split dataset to X and Y
```

```python
X = df.iloc[: , :-1]
y = df.iloc[: , -1]
```

```python
# Train test split
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X, y , test_size= .3 , random_state= 42 )
```

```python
# Model import and train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Regressore = LinearRegression()
```

```python
Regressore.fit(X_train , y_train)
```

```python
y_preds = Regressore.predict(X_test)
y_preds
```

```python
# Create R - squire and equation line.
```

```python
slr_score = Regressore.score(X_test , y_test)
slr_score
```

```python
coif = Regressore.coef_
coif
```

```python
interc = Regressore.intercept_
interc
```

```python
from sklearn.metrics import mean_squared_error
import math

rmse = math.sqrt(mean_squared_error(y_test , y_preds))

rmse
```
